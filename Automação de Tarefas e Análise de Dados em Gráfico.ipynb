```python
import yfinance
import pyautogui
import pyperclip
```

```python
# usado a funcionalidade Ticker para busca do código de cotação em tempo real
# em inglês, 6mo == 6 months == 6 meses.
ticker = input("Digite o código da cotação: ")
dados_nubank_6mo = yfinance.Ticker(ticker).history("6mo")
fechamento = dados_nubank_6mo.Close
# com a função PLOT é gerado o gráfico com os dados buscados
fechamento.plot()
```

# Análise de dados
### Buscando cotação máxima, mínima e atual

```python
maxima = fechamento.max()
maxima
```

```python
minima = fechamento.min()
maxima
```

```python
atual = fechamento[-1]
atual
```

# Automação - Envio de E-mail

```python
# COM PYAUTOGUI.HOTKEY é possível controlar seu teclado. Com PYAUTOGUI.CLICK é possível controlar o mouse.
# Com PYPERCLIP.COPY é possível copiar texto.

pyautogui.hotkey("ctrl", "t")
site = "www.gmail.com"
pyperclip.copy(site)
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")
# PYAUTOGUI.PAUSE realizar um delay de alguns segundos para os proximos comando. Isso é bom se seu sistema não é tão rápido.
pyautogui.PAUSE = 3
pyautogui.click(x=83, y=198)
email = "brunna-bruna@hotmail.com, roginhovs@gmail.com"
pyperclip.copy(email)
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")
assunto = "Análise de dados - NUBANK"
pyperclip.copy(assunto)
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")
# Ao usar o F no começo é possível usar variáveis dentro de string.
# Três aspas no começo e no fim para realizar quebra de texto.
# O uso da ENTRE CHAVES{} permite o correto uso de variáveis e outras funcionalidades dentro do texto.
# O uso da função ROUND serve para arredondar uma numeração. o número em seguida é o numero de casas decimais após a virgula.
mensagem = f"""
Prezada Gestora, 

Segue a análise de dados sobre a cotação das ações do NUBANK - {ticker}:

Cotação Máxima: R$ {round(maxima, 2)}
Cotação Mínima: R$ {round(minima, 2)}
Cotação Atual: R$ {round(atual, 2)}

"""
    
pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")
pyautogui.click(x=837, y=688)
```

```python
# O PYAUTOGUI.POSITION é usado para dar as coordenadas x e y da seta do mouse.
# Ele é usado para passar as coordenadas para usar no PYAUTOGUI.CLICK
pyautogui.position()
```
